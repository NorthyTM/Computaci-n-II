```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --- Dataframe de 700 registros ---
estadistica = pd.read_csv("vehiculos_700_registros.csv")

# Mostrar TODO el dataframe
pd.set_option('display.max_rows', None)
print(estadistica)

# --- Estadística descriptiva básica ---
print(estadistica.info())
print(estadistica.head())
print(estadistica.describe())
print(estadistica.shape)
print(estadistica.dtypes)

# Selección correcta de columnas
col_nombre = estadistica["Nombre_Vehiculo"]
col_precio = estadistica["Precio"]
col_ventas = estadistica["Ventas_Vehiculo"]

# Estadísticas
print(f"\nPromedio de ventas: {col_ventas.mean()}")
print(f"Valor mínimo de ventas: {col_ventas.min()}")
print(f"Valor máximo de ventas: {col_ventas.max()}")
print(f"Desviación estándar (Ventas): {col_ventas.std()}")
print(f"Desviación estándar (Precio): {col_precio.std()}")

# --- Gráfico de Barras ---
plt.figure(figsize=(12, 6))
plt.bar(col_nombre.head(10), col_precio.head(10))
plt.title("Precio de Vehículos (Top 10)")
plt.xlabel("Nombre del Vehículo")
plt.ylabel("Precio")
plt.xticks(rotation=45)
plt.show()

# --- Histograma ---
plt.figure(figsize=(10, 5))
plt.hist(col_ventas, bins=20)
plt.title("Distribución en Histograma de Ventas")
plt.xlabel("Rango de Ventas")
plt.ylabel("Frecuencia")
plt.show()
```
