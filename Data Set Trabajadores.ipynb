```python
import pandas as pd
import numpy as np
import random

idprofesor = [random.randint(10**10, 10**11 - 1) for _ in range(10)]

profesores = {
    "ID_Profesor": idprofesor,
    "Nombre": [np.nan,"Isaías", "Jorge", "Sebastián", "Guillermo", "Kevin", "Neftalí", "Gerson", np.nan, "Bernardo"],
    "Apellido": [np.nan, "Sandoval", "Monge", np.nan, "Moreno", "Hernández", "Herrera", "Melgar", "Pineda", "Méndez"],
    "Curso": ["Biología", "Química", "Seminario", np.nan, "Matemáticas", "Física Fundamental", np.nan, "Programación", "Historia", "Geografía"],
    "Edad": np.random.randint(20, 61, size=10),   
    "Sueldo": np.random.randint(3000, 5001, size=10)
}

profesores = pd.DataFrame(profesores)
print(profesores.isnull())
print(profesores.isnull().sum())

profesoresisna = pd.DataFrame(profesores).isna()
print(profesoresisna)

profesoresnull = pd.DataFrame(profesores).notnull()
print(profesoresnull)

profesoresfillna = pd.DataFrame(profesores).fillna("NoDatos")
print(profesoresfillna)

print(profesores)
```
